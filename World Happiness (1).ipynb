## Removing Outliers  

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df))
z.shape
```

```python
thresold =3
print(np.where(z>3))
```

```python
df_new =df[(z<3).all(axis=1)]
df_new.shape
```

## Percentage of data loss 

```python
p_loss= (158-149)/158*100
p_loss
```

```python
x=df_new.drop(columns=["Happiness Score"])
y=df_new["Happiness Score"]
```

## Remove Skewness 

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
x
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=scale.fit_transform(x)
```

## As the target has continue data, I will use KNN Regression model 

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
lr= LinearRegression()
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.25, random_state=25)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
print("R2 Score" ,r2_score(pred, y_test))
```

```python
from sklearn.model_selection import GridSearchCV
LinearRegression()
param_grid = {"algorithm":['kd_tree', 'brute'],
            'leaf_size':range(2,20),
            'n_neighbors':range(3,8)}
gridsearch = GridSearchCV(estimator=knn, param_grid=param_grid)
gridsearch.fit(x_train,y_train)
```

```python
gridsearch.best_params_
```

```python
knn = KNeighborsRegressor(algorithm='kd_tree', leaf_size=6, n_neighbors=6)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_train,y_train)
```

```python
pred = knn.predict(x_test)
r2_score(pred, y_test)
```
